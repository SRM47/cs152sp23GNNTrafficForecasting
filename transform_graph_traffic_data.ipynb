```python
import networkx as nx
import pylab as plt
```

1. Create new vertices (from the edges)
2. for each edge (u, v) in the original graph
       a. for all neighbours n_i of v,
              add new edge in the new graph between the vertez that represents the original edge and the vertex that represents the edge v, n_i

```python
G = nx.DiGraph()
nodes = ["A", "B", "C", "D", "E", "F"]
for node in nodes: 
       G.add_node(node)
G.add_edge("A", "B", weight=1)
G.add_edge("B", "C", weight=2)
G.add_edge("D", "C", weight=4)
G.add_edge("F", "A", weight=9)
G.add_edge("B", "E", weight=3)
G.add_edge("A", "D", weight=3)
```

```python
G_prime = nx.DiGraph()
edges = {}
for u, v in G.edges:
       w = G.get_edge_data(u, v)
       # create a new vertex in the new graph
       # each new node in the new graph will carry the edge information 
       G_prime.add_node(f"{u}->{v}", traffic_flow = w["weight"])
       edges[f"{u}->{v}"] = (u,v)

for u, v in G.edges:
       # find all neighbouring nodes
       for n in G.neighbors(v):
              # connect vertex (u,v) to (v,n) in the new graph
              G_prime.add_edge(f"{u}->{v}", f"{v}->{n}")
    
```

```python
nx.draw(G, with_labels = True)
```

```python
nx.draw(G_prime, with_labels=True)
```
